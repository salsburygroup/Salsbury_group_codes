```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# dK9
tmp=pd.read_csv("/deac/salsburyGrp/wud18/md/TM/logistic_regression_Hbond/TM.csv", sep=',', header=0)
tmp2=tmp.T
header=tmp.columns[1:]
header2=[float(x) for x in header]
colnames=tmp.iloc[:,0]
```

```python
df=pd.DataFrame(tmp2[1:len(tmp2)], columns = colnames)
df=tmp2[1:len(tmp2)]
```

```python
# Accuracy and max value
Accuracy=np.array(df.values)
print(np.max(Accuracy))
```

```python
# argmax
Argmax=np.unravel_index(Accuracy.argmax(), Accuracy.shape)
print("Accuracy=")
print(Accuracy[Argmax])
print('lambda=')
print(header[Argmax[0]])
print('alpha=')
print(colnames[Argmax[1]])
```

```python
fig1=plt.figure(1,figsize=(16,10))
plt.plot(header2, df, linewidth=0.5, marker='x')
#plt.plot(header2, header2)
plt.xlabel("lambda")
plt.ylabel('Accuracy')
plt.title('Elastic net for dK9')
plt.legend(colnames)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
fig1.savefig(os.path.join('/deac/salsburyGrp/wud18/md/double_binding/hbond/elastic_net/dK9.png'), pad_inches=0.03, bbox_inches='tight', dpi=200)
fig1.savefig(os.path.join('/deac/salsburyGrp/wud18/md/double_binding/hbond/elastic_net/dK9.tiff'), pad_inches=0.03, bbox_inches='tight', dpi=600)
fig1.savefig(os.path.join('/deac/salsburyGrp/wud18/md/double_binding/hbond/elastic_net/dK9.pdf'), bbox_inches='tight')
```
